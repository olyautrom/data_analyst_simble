# Портрет пользователя

## Описание

**Классификация пользователей по методологии RFM**

Методология RFM вполне простая: исследователю нужно присвоить каждому заказчику степень «1», «2», «3» в зависимости от:
- общей суммы отгруженных заказов (предоставленных и оплаченных услуг) за весь период работы с контрагентом. За нее отвечает показатель Monetary (M)

- частоты заказов – Frequency (F)

- насколько недавно была совершена последняя сделка – Recency (R)

R=3, F=1, M=1 = заказчики с самым высоким LTV и высокой частотой заказов, которые делали последний заказ давно, находятся в оттоке. Такие контрагенты требуют экстренных мероприятий по выявлению причин оттока и попыток вернуть клиента

M=1, F=1, R=1 = потенциально опинион-лидеры, адвокаты вашего бренда. Также полезно изучить качественный состав таких контрагентов при большом количестве сегментов целевой аудитории в вашей базе

M=2/3, F=1, R=1= в таком сегменте здорово проводить мероприятия, направленные на увеличение среднего чека

RFM-анализ местами перекликается с известным Законом Парето, гласящим, что 80% результатов происходят благодаря 20% усилий. Когда на этот закон смотрят через призму маркетинга, это означает, что 80% всех ваших продаж исходят от 20% наиболее лояльных и постоянных клиентов.

## Расчеты RMF

### Frequency

Соберем частоту закзазов по пользователю: Frequency 

```python
frequency = data.groupby('user_id').agg({'policy_id': 'count'}).reset_index().rename(
    columns={'policy_id': 'number_policies'})
frequency
```

Выведем распределение частоты методом describe

```python
frequency.describe(percentiles = [.2, .4, .6, .8, .9, .99]).round()
```

- 80% пользователей совершали покупку 2 и менее раз
- 90% пользователей совершали покупку 4 и менее раз
- количество покупок на пользователя от 1 до 39

### Monetary

Найдем общую сумму заказов по каждому пользователю (LTV по общему доходу):

```python
monetary = data.groupby('user_id').agg({'cost_of_policy': 'sum'}).reset_index().rename(
    columns={'cost_of_policy': 'amount'})
monetary
```

```python
monetary.describe(percentiles = [.2, .4, .6, .8, .9, .99]).round()
```

В среднем выручка с пользователя 1790р. Разброс от 5 до 13275р

### Recency

Найдем даты последних покупок по пользователю, чтобы описать показатель Recency:

```python
recency = data.groupby('user_id').agg({'date': 'max'}).reset_index().rename(
    columns={'date': 'last_policy_date'})
recency
```

Расчитаем число дней с последней покупки до текущей даты

```python
# до текущей даты
# recency['current_date'] = pd.to_datetime(datetime.date.today())
recency['current_date'] = pd.to_datetime('2023-09-25')
recency['churn_time'] = recency['current_date'] - recency['last_policy_date']
recency
```

```python
recency.describe(percentiles = [.2, .4, .6, .8, .9, .99])
```

- Среднее время с последней покупки 56 дней
- У 40% пользователей последняя покупка была меньше, чем 22 дня назад
- у 60% пользователей последняя покупка была меньше, чем 48 дней назад
- у 80% пользователей последняя покупка была меньше, чем 84 дня назад
- Минимальное число дней с последней покупки 0, максимальное - 264

### Объдинение таблиц

Объединим таблицы `frequency`, `monetary` и `recency`

```python
rmf = frequency.merge(
    monetary, left_on='user_id', right_on='user_id', how="outer").merge(
    recency[['user_id', 'churn_time']], left_on='user_id', right_on='user_id', how="outer"
)
rmf
```

### Расчет степеней R, M, F

Присвоим степени 1, 2, 3 по каждому свойству

По свойству частоты покупок `Frequency` выберем резделители: больше 1 покупки - 80 процентиль, больше 4 покупок - это 90 процентиль.

По сумме выручки `Monetary` границы степеней выберем: 1083р (чуть больше тысячи) - 50 процентиль, 2573р - 80 процентиль.

По числу дней с последнего полиса `Recency` границы поставим по 33 дням (месяц) - 50 процентиль и 84 дня (3 месяца) - 80 процентиль

```python
# создание столбцов 
rmf[['R', 'M', 'F']] = ''
# степени F
rmf.loc[rmf['number_policies'] > rmf['number_policies'].quantile(0.9), 'F'] = '1'
rmf.loc[rmf['number_policies'] <= rmf['number_policies'].quantile(0.9), 'F'] = '2'
rmf.loc[rmf['number_policies'] <= rmf['number_policies'].quantile(0.8), 'F'] = '3'
# степени M
rmf.loc[rmf['amount'] > rmf['amount'].quantile(0.8), 'M'] = '1'
rmf.loc[rmf['amount'] <= rmf['amount'].quantile(0.8), 'M'] = '2'
rmf.loc[rmf['amount'] <= rmf['amount'].quantile(0.5), 'M'] = '3'
# степени R
rmf.loc[rmf['churn_time'] > rmf['churn_time'].quantile(0.8), 'R'] = '3'
rmf.loc[rmf['churn_time'] <= rmf['churn_time'].quantile(0.8), 'R'] = '2'
rmf.loc[rmf['churn_time'] <= rmf['churn_time'].quantile(0.5), 'R'] = '1'
# вывод таблицы
rmf.head(10)
```

```python
# оставляем тольку нужные столбцы
rmf = rmf[['user_id', 'R', 'M', 'F']]
rmf
```

## Выделение сегментов

### Сегмент attention

Сегмент attention требует изучения причин оттока, одни из самых крупных клиентов с самыми старыми датами последних заказов, которых мы найдем так:

`M == "1" & R == "3"`

```python
attention = rmf.query('M == "1" & R == "3"').reset_index(drop=True)
attention
```

Таких пользователей 8. Напишем функию для вывода сегметнов и отобразим сегмент attention:

```python
def view_segment(segment):
    

    current_list = list(segment['user_id'])

    segment_table = data.query('user_id in @current_list').groupby('user_id').agg({'surname': 'max', 
                                 'policy_id': 'count', 'user_source': 'max', 'cost_of_policy': 'sum',
                                 'vehicle': 'max', 'date': 'max'}).reset_index()
    segment_table.columns = ['user_id', 'surname', 'num_policies', 'user_source', 'amount', 
                               'vehicle', 'last_policy_date']
    return segment_table

attention_table = view_segment(attention)
attention_table
```

Есть пользователи, которые совершали по одной дорогой покупке из-за авто или большого пакета. Или несколько покупок  совокупно на значимую сумму. По сути это все ушедшие клиенты, которые были значимы

### Сегмент лидеров

Или найдем самых классных, верных и упитанных пользователей, чтобы выписать им промокод на полис, свозить на море и прочее:

```python
top = rmf.query('M == "1" & R == "1" & F == "1"').reset_index(drop=True)
top
```

Таких пользователей 12, выведем сегмент лидеров

```python
top_table = view_segment(top)
top_table
```

Постоянные пользователи, часто покупающие со значимой суммой покупок

- Пользоватеи с наимбольшим числом полисов - это Бузоверов, Ненюков, Шейкман

- Пользователи с наибольшей суммой покупок - это Чукалин, Петрова, Шейкман, Карпенков



### Сегмент потенциала

Последний сегмент, неактивные пользователи, которых нужно вовлекать в дальнейшие покупки, с показателями `M=2/3, F=1, R=1`

```python
potential = rmf.query('M != "1" & R == "1" & F == "1"').reset_index(drop=True)
potential
```

Таких пользователей 9, отобразим сегмент потенциала:

```python
potential_table = view_segment(potential)
potential_table
```

У этих пользователей выручка меньше, но частота покупок высокая и совершали они их не так давно. Выведем тарифы, которые приобритают этот сегмент

```python
potential_list = list(potential["user_id"])
data.query('user_id in @potential_list').groupby('type').agg({'policy_id': 'count'}).reset_index()
```

Этот сегмент основывается на парковках, что логично, частые покупки с низким чеком
